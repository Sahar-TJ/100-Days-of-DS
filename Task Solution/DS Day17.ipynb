## Pandas: Visualization

__Data visualization is the practice of translating information into a visual context, such as a map or graph, to make data easier for the human brain to understand and pull insights from. The main goal of data visualization is to make it easier to identify patterns, trends and outliers in large data sets. Matplotlib is a Library that is used to visualize, it is mostly already imported__

```python
import pandas as pd
import numpy as np

df = pd.read_csv(r'C:\Users\user\Desktop\100daysds\class_grade.csv')
df.info()
```

__Line Plot__

```python
df.plot(figsize=(15,5))
```

__Bar Plot__

```python
df.plot(kind="bar",x= "Final",figsize=(30,10))
```

__Stacked Bar Plot__

```python
df.plot.bar(stacked=True, figsize=(25,10))
```

__Horizontal Stacked Bar Plot__

```python
df.plot.barh(stacked=True)
```

__Histogram Graph__

```python
df.plot.hist()
```

__Stacked Histogram Plot__

```python
df.plot.hist(stacked=True, bins =20)
```

__Cumulative Horizontal Histogram__

```python
df.plot.hist(orientation="horizontal", cumulative=True)
```

__Histogram of each column in the Dataframe__

```python
df.diff().hist()
```

__Area Plot__

```python
df.plot.area()
```

__Box Plot__

```python
df.plot.box(figsize=(25,8))
```

__Horizontal Box Plot__

```python
df.plot.box(vert=False)
```

__Scatter Plot__

```python
df.plot.scatter("TakeHome", "Final")
```

__Plotting 2 different scatter plots in one__

```python
ax=df.plot.scatter(x="TakeHome", y="Final", color="red", marker="*", s=100)
df.plot.scatter(x='Tutorial', y='Assignment', grid=True, ax=ax, s=25)
```

__Pie Chart__

```python
df['Grade'].plot.pie()
```

__Since the pie chart above has too many pieces, we will group them as one and then plot.__

```python
ad= df['Grade'].value_counts()
ad.plot.pie()
```

__Density Plot__

```python
df.plot.kde(figsize=(25,10))
```
