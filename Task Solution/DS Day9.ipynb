## Pandas: Indexing, Selection and Assignment


__Indexing:__ Indexing in pandas means simply selecting particular rows and columns of data from a DataFrame.  
It is done by using .loc[], .iloc[ ]
- df.[ ] : This function also known as indexing operator
- df.loc[ ] : This function is used for labels.
- df.iloc[ ] : This function is used for positions or integer based

__Slicing:__ Slicing is when we select a set of data from the DataFrame.
It is done by using ' : ' symbol.

```python
import pandas as pd
data = pd.read_csv("nba.csv", index_col ="Name")
data.info()
```

### Indexing

### df.[]

```python
ind = data[["Team", "College"]]
ind
```

### .loc[]

```python
# Selecting a row and displaying it's info
ri1 = data.loc["Avery Bradley"]
print(ri1)
ri2 = data.loc["R.J. Hunter"]
print(ri2)

#Selecting 2 rows and displaying their info
ri3 = data.loc[["Jae Crowder", "Jonas Jerebko"]]
ri3
```

```python
#selecting 2 rows and 3 columns
ri4 = data.loc[["Avery Bradley", "R.J. Hunter"],["Team", "Number", "Position"]]
ri4
```

```python
# Selecting only the columns
ri5 = data.loc[:, ["Team", "Number", "Position"]]
ri5
```

### .iloc[]

```python
row1 = data.iloc[4]   
print(row1)

row2 = data.iloc[['2','5','7']]
row2
```

```python
# Selecting 2 rows and 2 columns
row3 = data.iloc [[3, 4], [1, 2]]
row3
```

```python
#Selecting all the rows and 3 columns
row4 = data.iloc[:,[3,4,6]]
row4
```

### Slicing

```python
data[1:3]
```

```python
data[-3:]
```
