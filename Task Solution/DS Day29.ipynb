## Stats - Central Limit Theorem and Margin of Error

__Central Limit Theorem (CLT):__ The sample averages/mean will tend towards a normal distribution regardless of the population distribution.

__Importing Libraries and Data__

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid')
```

```python
data = pd.read_csv(r'C:\Users\user\Desktop\100daysds\clt_data.csv')
print(data.head())
```

__Distribution__

```python
plt.hist(data.Weight, density=True, bins=30) 
```

```python
def random_subsets(data, tot_subsets, sub_size):
    subset_means = []
    for i in range(tot_subsets):
        subset = data.sample(n=sub_size)
        subset_mean = np.array(subset).mean()
        subset_means.append(subset_mean)
    return subset_means
```

```python
subsets_from_data = random_subsets(data, 100, 50) 
# Keep increasing the total number of subsets
sns.distplot(subsets_from_data, hist=False)
```

```python
data2 = pd.read_csv(r'C:\Users\user\Desktop\100daysds\clt_data2.csv')
plt.hist(data2.Weight, density=True, bins=30) 
```

```python
subsets_from_data2 = random_subsets(data2, 5000, 30) 
sns.distplot(subsets_from_data2, hist=False)
```

__Margin of Error:__ Measure the maximum amount by which the sample results are expected to differ from those of the actual population. Lower margin of error indicates higher confidence levels in the produced results.


where,
p = sample proportion (“P-hat”).

n = sample size

z = z-score corresponds to your desired confidence levels.

We can use the following formulas for MoE:
- Standard deviation from the population multiplied by the critical value = margin of error
- Standard error of the sample multiplied by the critical value = margin of error

__Example:__
A publishing company wants to conduct a survey of its customers to see whether they prefer to read eBook or physical books. The company has roughly one million customers in its database. Because the company cannot realistically survey all one million individuals, they gather a random sample from 3,000 customers that represent the full customer base.

After the completion of the survey, the data shows that 2,000 out of 3,000 customers prefer physical books __(2000/3000 = .67 or 67%)__. Using a __level of confidence of 95%__, we can determine that the __margin of error will be 2%__. This means that anything between __65% and 69%__ could be considered accurate in reference to how many customers prefer physical books out of the entire customer database. The company can then assume that if they were to continue to survey the remaining population of their customer base, they would find that 65% to 69% of individuals would rather read a physical book than an eBook.
