## Pandas: Rename and Replace

__Rename :__ Changing the name of a column or a row index.

__Replace :__ Changing an already given value to a new value.

```python
import pandas as pd
import numpy as np

data = pd.read_csv(r"C:\Users\user\Desktop\100daysds\ST-Data.csv")
data.info()
```

### Renaming

__Syntax:__ *DataFrame.rename(mapper=None, index=None, columns=None, axis=None, copy=True, inplace=False, level=None, errors='ignore')*

__We will first check the column names of the dataset__

```python
data.columns
```

__Renaming Single Column__

```python
data.rename(columns = {'Data Science: (Worldwide)': 'Data Science'}, inplace = True)
data.head()
```

__Renaming multiple columns__

```python
data.rename(columns = {'Machine learning: (Worldwide)' : 'Machine Learning', 
                       'Artificial intelligence: (Worldwide)' : 'Artificial Intelligence'},
                       inplace = True)
data.head()
```

__Another way of Renaming Columns is by already giving them a name__

```python
new_cols = ['Date', 'Data Science', 'Machine Learning', 'Artificial Intelligence']

data.columns = new_cols
data.head()
```

__Renaming Columns while importing the data__

```python
new_cols = ['Date', 'Data Science', 'Machine Learning', 'Artificial Intelligence']

df = pd.read_csv(r"C:\Users\user\Desktop\100daysds\ST-Data.csv", names = new_cols)
df.head()
```

```python
df.drop(0, axis = 0, inplace=True)
df.head()
```

__Perform various functions by renaming columns__

```python
df.rename(columns = str.lower)
df.head()
```

__Renaming Indexes__

```python
df.rename(index={1: 111, 2: 222, 3: 333}, inplace = True)
df.head()
```

__Renaming row indexes or column index or both__

```python
df1 = df.rename_axis("Row Number", axis='rows')
df1.head()
```

```python
df2 = df.rename_axis("Topics", axis='columns')
df2.head()
```

```python
df3 =df.rename_axis("Row", axis='rows').rename_axis("Topics", axis='columns')
df3.head()
```

### Replacing

__Syntax:__ *DataFrame.replace(to_replace=None, value=None, inplace=False, limit=None, regex=False, method='pad')*

__Let's just replace the value of Artificial Intelligence in the first row__

```python
df.replace(to_replace ="61", value ="67")
df.head()
```

__Replacing multiple values with a single value__

```python
df.replace(to_replace =["57", "58"], value ="55")
```

__Replacing multiple values with multiple values__

```python
df.replace(to_replace =["5", "4"], value =["7","6"], inplace = True)
df.head(10)
```

__Replacing Null values__

```python
import numpy as np

df.loc[-1] = [ np.nan  , np.nan, 4, np.nan]  # adding a row with Null values
df.index = df.index + 1 
#df = df.sort_index(ascending=False) 

df.tail(10)
```

```python
df.replace(np.nan, -999, inplace = True)
df.tail(10)
```

__Note:__ The inplace is by default set to be false.
