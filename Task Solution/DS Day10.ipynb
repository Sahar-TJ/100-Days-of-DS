## Pandas: Iteration and Sorting

__Iteration:__ Iteration is a general term for taking each item of something, one after another. Pandas DataFrame consists of rows and columns so, in order to iterate over dataframe, we have to iterate a dataframe like a dictionary. There are 3 ways to do it, __.items(), .iterrows(), .itertuples()__

__Sorting:__ Sorting is arranging the dataframe in an order. It is done by __.sort_values()__ and __.sort_index()__

```python
import pandas as pd
dat = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv')
#Since the data is huge I am just randomly extracting data.
df = dat.sample(500)
df.info
```

### Iteration

### df.items()

```python
for col_name, data in df.items():
    print("Column name: ", col_name, "\nData: ", data)
```

__Iterating only a specific data item__

```python
for col_name, data in df.items():
    print("col_name:",col_name, "\ndata:",data[1])
```

### df.iterrows()

```python
for index, row in df.iterrows():
   print(index, "carat: ", row.carat, "cut: ", row.cut,"color: ", row.color,"price: ", row.price)
```

```python
for i, row in df.iterrows():
    print(f"Index: {i}",f"{row}\n")
    #print()
```

__Iterating a specific data item using .iterrows()__

```python
for i, row in df.iterrows():
    print(i, row[3])
```

### .itertuples()

```python
for row in df.itertuples():
    print(row)
```

__The itertuples() method has two arguments: index and name.
We can choose not to display index column by setting the index parameter to False__

```python
for row in df.itertuples(index= False):
    print(row)
```

__As it's seen there is Pandas written in each Tuple, the generator yields namedtuples with the default name of Pandas. We can change the name parameter.__

```python
for row in df.itertuples(index=False, name='Diamonds'):
    print(row)
```

### Sorting

### .sort_values()

```python
df.sort_values("price")
```

__Sorting data by a specific column and in descending order__

```python
df.sort_values(by="depth",ascending=False)
```

### Sorting Algorithm
__Its good to note that pandas allows you to choose different sorting algorithms to use with both .sort_values() and .sort_index(). The available algorithms are quicksort, mergesort, and heapsort. quicksort is the default sorting algo, mergesort is the stable algo, good for multiple sorting__

```python
df.sort_values( by="cut", ascending=False, kind="mergesort" )
```

__Sorting by Multiple Columns in Ascending Order__

```python
df.sort_values(by=["cut", "price"])[["cut", "price"]]
```

__Changing the Column Sort Order__

```python
df.sort_values(by=["price", "cut"])[["cut", "price"]]
```

__Sorting by Multiple Columns in Descending Order__

```python
df.sort_values(by=["cut", "price"], ascending = False)[["cut", "price"]]
```

__Sorting by Multiple Columns With Different Sort Orders__

```python
df.sort_values( by=["cut", "color", "price"], ascending=[True, True, False])[["cut", "color", "price"]]
```

### .sort_index()
A DataFrame has an .index property, which by default is a numerical representation of its rows locations. You can think of the index as the row numbers. It helps in quick row lookup and identification.

```python
sorted_df = df.sort_values(by=["color", "clarity"])
sorted_df
```

```python
sorted_df.sort_index()
```

__Setting a custom index by using .set_index()__

```python
assigned_index_df = df.set_index(["color", "cut"] )
assigned_index_df
```

```python
assigned_index_df.sort_index()
```

__Sorting by Index in Descending Order__

```python
assigned_index_df.sort_index(ascending = False)
```

### Sorting the Columns

__When you use .sort_index() without passing any explicit arguments, it uses axis=0 as a default argument. The axis of a DataFrame refers to either the index (axis=0) or the columns (axis=1). __

```python
df.sort_index(axis=1)
```

```python
df.sort_index(axis=1, ascending = False)
```

### Using .sort_values() and .sort_index() In Place
__With inplace set to True, you modify the original DataFrame, so the sort methods return None.__

```python
df.sort_values("cut", inplace=True)
```

```python
df.sort_index(inplace=True)
df
```

__Note:__ We can also use __na_position="first"__ Parameter in .sort_values() and .sort_index() to display the missing values first. By default it will be in last.
