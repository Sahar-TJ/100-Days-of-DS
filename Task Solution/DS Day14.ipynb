## Pandas: Combine, Merge and Join

__Combine :__ for combining DataFrames across rows or columns. Combines any sort of datasets together.

__Merge :__ for combining data on common columns or indices. Will only merge similar data items from both the datasets

__Join :__ for combining data on a key column or an index. By default Joins the tables by left join.

```python
import pandas as pd
import numpy as np

data = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv")
#Just divided the dataset into 2 different datasets
df = data.sample(100)
df2 = data.sample(100)

print(df.info())
print(df2.info())
```

```python
df2 = df.drop(["carat", "color", "clarity"], axis=1)
df2
```

### Combining

It will just join 2 different dataframes into one without checking if there are any similar data items

```python
new_df = pd.concat([df, df2])
new_df
```

### Joining

```python
new_df2 = df.join(df2, lsuffix = '_LEFT', rsuffix= '_RIGHT')
new_df2
```

__Joining using a common column__

```python
new_df3 = df.set_index('cut').join(df2.set_index('cut'),lsuffix = '_LEFT', rsuffix= '_RIGHT')
new_df3
```

__Joining by preserving a specific column of the first dataframe__

```python
new_df4 = df.join(df2.set_index('cut'),on = 'cut' ,lsuffix = '_LEFT', rsuffix= '_RIGHT')
new_df4
```

### Merge

__Merge is by default Left Join__

```python
new_df5 = pd.merge(df, df2)
new_df5
```

__Right Join__

```python
new_df6 = pd.merge(df, df2, on='depth', how='right')
new_df6
```

__Outer Join__

```python
new_df7 = pd.merge(df, df2, on='depth', how='outer')
new_df7
```

__Inner Join__

```python
new_df8 = pd.merge(df, df2, how='inner')
new_df8
```

__Joining using multiple columns__

```python
new_df9 = pd.merge(df, df2, on=['depth', 'cut'], how='left')
new_df9
```

__Note:__ We can even use append() in case of concat(). Can be used both on Series and Dataframes.
