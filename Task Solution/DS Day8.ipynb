## Introduction to Pandas, Read and Write Data Frame


__Panda:__ It is a Library used for data manipulation and analysis.  Pandas is fast and it has high performance & productivity for users.
To get started with Pandas, all we need to do is import the library by using __import pandas__

### Commonly Used Functions

#### Reading a file

```python
import pandas as pd
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data')
iris.head()
```

#### Reading the File without the header

```python
irishead = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data", header=None)
irishead.head()
```

#### Giving Custom names to the columns in the dataset while Reading

```python
irisc = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",
                 names = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm', 'Species'])
irisc.head()
```

#### Displaying the rows

```python
iris.head(10)
```

#### Displaying random rows

```python
irishead.sample(10)
```

#### Displaying the last few rows of the dataset

```python
iris.tail(15)
```

#### Displaying the whole dataset

```python
print(iris)
```

#### To find out the number of columns and rows in the Dataset

```python
irisc.shape
```

#### To know more about the dataset

```python
irishead.info()
```

#### To find out if there are null items in the Dataset

```python
iris.isnull().sum()
```

#### Counting the diffirent categories in the column

```python
irishead[4].value_counts()
```

#### Deleting a column or two
We provide “inplace=True’ to modify the current DataFrame.

```python
irisc.drop([0], inplace = True)
irisc.head()
```

#### To find out various numerical functions in the dataset, like mean, standard deviation, etc.

```python
iris.describe()
```

#### Checking the number of unique values in the Columns

```python
iris['5.1'].nunique()
```

#### Using the nsmallest function

“nsmallest() & nlargest()” functions are used to obtain “n” no. of rows from our dataset which are lowest or highest respectively.

```python
irishead.nsmallest(5,1)
```

#### Using the nlargest funtion

```python
irishead.nlargest(5, 1)
```

#### Grouping the dataset and counting the number of similar values

```python
iris.groupby('5.1').agg('count')
```

#### Grouping the dataset with a certain category
We can use the “get_group()” function to select a specific group.

```python
irishead.groupby(4).get_group('Iris-virginica')
```

#### Sorting values 
Can use .sort_index or .sort_values 

```python
irishead[[0,1,2,3]].sort_values(by = 2)
```

#### Using the selct_dtype() to extract the columns that have certain type of values 

```python
numiris = irishead.select_dtypes(include ='number')
numiris.head()
```

```python
catiris = irishead.select_dtypes(include = 'object')
catiris.head()
```

#### Concatinating 2 or more datasets into one
By default, axis=0, i.e. row-wise concatenation, so if we set axis=1, column-wise concatenation will be performed.

```python
combiris = pd.concat([numiris, catiris], axis=1)
combiris.head()
```

#### Changing categorical values to variables
Pandas “get_dummies()” method is used to convert the categorical features of the data into dummy variables or indicator variables.

```python
diris = pd.get_dummies("4", prefix = "4.0").iloc[:,:]
diris.head()    
```

#### loc() and iloc()
loc() and iloc() methods are used in slicing data from the pandas DataFrame which helps in filtering the data according to some given condition.

- loc – select by labels
- iloc – select by positions

```python
# Values till 3 of the columns 1 and 3 will be displayed
irishead.loc[:3,[1,3]]
```

```python
#iloc() slices the data frame in the specified rows and column range.
irishead.iloc[:5,:2]
```
