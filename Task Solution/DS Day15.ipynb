## Pandas: Summary, Crosstab and Pivot

__Summary:__ Used to describe the dataset

__Crosstab:__ Tells us if there is any pattern / relationship between attributes

__Pivot:__ Creates a pivot table of the dataframe

```python
import pandas as pd
import numpy as np

data = pd.read_csv(r"C:\Users\user\Desktop\100daysds\class_grade.csv")
data
```

### Summary

__Syntax:__ *DataFrame.describe(percentiles=None, include=None, exclude=None, datetime_is_numeric=False)*

```python
data.describe()
```

__Summary of a specific column__

```python
data.Grade.describe()
```

__Finding the Standard deviation of a column__

```python
data.Assignment.std()
```

__Counting the number of values of a column__

```python
data.Final.value_counts()
```

__Summary of the dataset with a certain type of data__

```python
data.describe(include=['int64'])
```

__Summary of the dataset which includes all the datatypes__

```python
data.describe(include='all')
```

### Crosstab

__Syntax:__ *pandas.crosstab(index, columns, values=None, rownames=None, colnames=None, aggfunc=None, margins=False, margins_name='All', dropna=True, normalize=False)*

__Simple crosstab where the relation of 2 or more columns is checked__

```python
pd.crosstab(data['Grade'], data['Final'])
```

__Transposing the Crosstab__

```python
pd.crosstab(data.Grade, data.Final).T
```

__Taking out the percentage of the Crosstab__

```python
pd.crosstab(data.Tutorial, data.Grade).apply(lambda r: (r/len(data))*100, axis=1)
```

### Pivot

__Syntax:__ *pandas.pivot_table(data, values=None, index=None, columns=None, aggfunc='mean', fill_value=None, margins=False, dropna=True, margins_name='All', observed=False, sort=True)*

__Creating a new table where the index and values are set__

```python
pd.pivot_table(data, index=['Assignment', 'TakeHome'], values='Grade', aggfunc='mean')
```

__Setting index and Columns of the Pivot table__

```python
pd.pivot_table(data, index='Grade', columns=['Tutorial', 'TakeHome'], aggfunc='max')
```

__Performing multiple functions in a Pivot Table__ 

```python
table = pd.pivot_table(data, values=['Midterm', 'Final'], index=['Grade'],
                    aggfunc={'Midterm': np.mean, 'Final': [min, max, np.mean]})
table
```
